```python
from sklearn.preprocessing import MultiLabelBinarizer
mlb=MultiLabelBinarizer()
cuisine=mlb.fit_transform(dataset["Cuisines"])
dataset=pd.concat([dataset,pd.DataFrame(cuisine,columns=mlb.classes_)], axis=1)
dataset
```

```python
dataset=dataset.drop(columns=["Cuisines","Switch to order menu"])
dataset.info()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(handle_unknown="ignore")
ohe_columns=["Has Table booking","Has Online delivery","Is delivering now"]
enco_fit=ohe.fit_transform(dataset[ohe_columns])
encoded_df = pd.DataFrame(enco_fit.toarray(), columns=ohe.get_feature_names_out())
dataset=pd.concat([dataset,encoded_df],axis=1)
dataset=dataset.drop(columns=ohe_columns)
dataset
dataset.info()
```

```python
Rating_text_map={'Excellent':6,'Very Good':5,'Good':4,'Average':3,'Not rated':2,'Poor':1}
```

```python
dataset=dataset.drop(columns="Rating color")
```

```python
dataset["Rating text"]=dataset["Rating text"].map(Rating_text_map)
dataset.info()
```

```python
import seaborn as sns
plt.figure(figsize=(100,100))
sns.heatmap(dataset.corr(),annot=True,cmap="Reds")
plt.show()
```

```python
X=dataset.drop(columns=["Restaurant ID","Aggregate rating"])
Y=dataset["Aggregate rating"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test = train_test_split(X,Y,random_state=42,test_size=0.25,shuffle=True)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
Result=model.predict(X_test)
mse=mean_squared_error(y_test,Result)
r2=r2_score(y_test,Result)
print(mse,r2)
```

```python
from sklearn.linear_model import SGDRegressor
from sklearn.svm import SVR
model3=SGDRegressor()
model3.fit(X_train,y_train)
Y_pred3=model3.predict(X_test)
mse=mean_squared_error(y_test,Y_pred3)
r2=r2_score(y_test,Y_pred3)
print(mse,r2)
```

```python
model4=SVR()
model4.fit(X_train,y_train)
Y_pred4=model4.predict(X_test)
mse=mean_squared_error(y_test,Y_pred4)
r2=r2_score(y_test,Y_pred4)
print(mse,r2)
```

```python
from sklearn.tree import DecisionTreeRegressor
model5=DecisionTreeRegressor(random_state=0)
model5.fit(X_train,y_train)
Y_pred5=model5.predict(X_test)
mse=mean_squared_error(y_test,Y_pred5)
r2=r2_score(y_test,Y_pred5)
print(mse,r2)
```

Decision tree regressor is best
