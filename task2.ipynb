```python
Cuisine_type=Cuisine[list].stack()
unique_cuisine=Cuisine_type.unique()
unique_cuisine
```

```python
z=len(unique_cuisine)
unique_cuisine[0]
```

```python
x={}
n=0
for k in range(0,z):
    if unique_cuisine[k]==0:
        n=1
    if n==0:
        x[unique_cuisine[k]]=k+1
    else:
        x[unique_cuisine[k]]=k
x
```

```python
dataset=dataset.drop(columns=["Restaurant Name","Country Code","City","Address","Locality","Locality Verbose","Longitude","Latitude","Currency","Rating color","Cuisines"])
dataset.info()
```

```python
Rating_text_map={'Excellent':6,'Very Good':5,'Good':4,'Average':3,'Not rated':2,'Poor':1}
dataset["Rating text"]=dataset["Rating text"].map(Rating_text_map)
dataset.info()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(handle_unknown="ignore")
ohe_columns=["Has Table booking","Has Online delivery","Is delivering now"]
enco_fit=ohe.fit_transform(dataset[ohe_columns])
encoded_df = pd.DataFrame(enco_fit.toarray(), columns=ohe.get_feature_names_out())
dataset=pd.concat([dataset,encoded_df],axis=1)
dataset=dataset.drop(columns=ohe_columns)
dataset
```

```python
dataset=dataset.drop(columns="Switch to order menu")
dataset.info()
```

```python
X=dataset.drop(columns=["Restaurant ID","Aggregate rating"])
Y=dataset["Aggregate rating"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test = train_test_split(X,Y,random_state=42,test_size=0.25,shuffle=True)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
Result=model.predict(X_test)
mse=mean_squared_error(y_test,Result)
r2=r2_score(y_test,Result)
print(mse,r2)
```

```python
from sklearn.tree import DecisionTreeRegressor
model5=DecisionTreeRegressor(random_state=0)
model5.fit(X_train,y_train)
Y_pred5=model5.predict(X_test)
mse=mean_squared_error(y_test,Y_pred5)
r2=r2_score(y_test,Y_pred5)
print(mse,r2)
```
